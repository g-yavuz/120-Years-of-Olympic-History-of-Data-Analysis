```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
df = sns.load_dataset("tips")
```

```python
df.info()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x="sex",y="tip",hue="day",legend=False)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x="sex",y="tip",hue="sex",legend=False)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x="day",y="tip",hue="sex",data=df)
plt.show()
```

```python
df["Tip Percentage"] = df["tip"] / df["total_bill"] * 100
```

```python
df
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df, x="day", y="total_bill", hue="smoker")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    data=df, 
    x="total_bill", 
    y="tip", 
    hue="size", 
    size="size", 
    sizes=(50, 250), 
    palette="viridis"
)
plt.show()
```

```python
df_weekend = df[df["day"].isin(["Sat","Sun"])]
df_weekend.groupby("day",observed=True)[["total_bill","Tip Percentage"]].describe()
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(data=df, x="time", y="Tip Percentage", hue="sex", split=True)
plt.show()
```

```python
sns.pairplot(df, hue="sex", palette="Set1")
plt.show()
```

```python
sns.jointplot(data=df, x="total_bill", y="tip", hue="sex", palette="Set1", height=7, s=70, alpha=0.7)

plt.show()
```
