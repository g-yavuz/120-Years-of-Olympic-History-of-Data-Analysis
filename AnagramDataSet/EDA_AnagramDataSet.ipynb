```python
diveded_people = df[df["attnr"]=="divided"]
```

```python
diveded_people["num1"].max()
```

```python
display(diveded_people[diveded_people["num1"] == diveded_people["num1"].max()])
display(diveded_people[diveded_people["num2"] == diveded_people["num2"].max()])
display(diveded_people[diveded_people["num3"] == diveded_people["num3"].max()])
```

```python
df["Max Score"]= diveded_people[["num1","num2","num3"]].sum(axis=1)
df
#It's just for the make somethings more difficult absolutly we can do like this df["Max Score"]= df[["num1","num2","num3"]].sum(axis=1)
```

```python
focused_people = df[df["attnr"]=="focused"]
focused_people
```

```python
df.loc[df["attnr"] == "focused", "Max Score"] = focused_people[["num1","num2","num3"]].sum(axis=1)
```

```python
df
```

```python
display(df.sort_values("Max Score",ascending=False).reset_index(drop=False))
df_new_sorted = df.sort_values("Max Score",ascending=False).reset_index(drop=True)

# if you write drop=True that means inplace=true
```

```python
df.groupby("attnr")[["num1","num2","num3"]].describe()
```

```python
plt.figure(figsize=(15,8))

plt.subplot(1,3,1)
sns.barplot(data=df,x="attnr",y="num1",hue="attnr", legend=False, errorbar=None, palette="viridis")
plt.title("First Task Success Based on Attention Level")

plt.subplot(1,3,2)
sns.barplot(data=df,x="attnr",y="num2",hue="attnr", legend=False, errorbar=None, palette="viridis")
plt.title("Second Task Success Based on Attention Level")

plt.subplot(1,3,3)
sns.barplot(data=df,x="attnr",y="num3",hue="attnr", legend=False, errorbar=None, palette="viridis")
plt.title("Third Task Success Based on Attention Level")

plt.show()
```

```python
plt.figure(figsize=(16,8))

plt.subplot(1,3,1)
sns.boxplot(data=df,x="attnr",y="num1",hue="attnr", legend=False, palette="Set2")
plt.title("Outliners")

plt.subplot(1,3,2)
sns.boxplot(data=df,x="attnr",y="num2",hue="attnr", legend=False, palette="Set2")
plt.title("Outliners")

plt.subplot(1,3,3)
sns.boxplot(data=df,x="attnr",y="num3",hue="attnr", legend=False, palette="Set2")
plt.title("Outliners")

plt.show()
```

```python
corr_map = df[["num1","num2","num3"]].corr()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(corr_map,annot=True, cmap="Blues", vmin=0, vmax=1)
plt.show
```

```python
table_pivot=df.pivot_table(
    values=["num1","num2","num3"],
    index="attnr",
    aggfunc="mean",observed=False
    
)
plt.figure(figsize=(10,6))
sns.heatmap(
    data=table_pivot,annot=True,cmap="Blues",
)
plt.show()
#Meaningless heatmap
```
